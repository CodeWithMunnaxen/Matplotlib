# Data Visualization
- Data Visualization is the process of representing data in a graphical or visual format such as charts, graphs, maps, and dashboards so that people can understand patterns, trends, and insights easily

# OR

- Data Visualization is the art of converting raw data into visual representations to make data easier to understand and analyze.

# Benefits:

- Easy to understand complex data
- Identifies patterns and trends
- Helps in decision making
- Detects outliers and anomalies
- Improves business insights
- Makes reports attractive

# Data Visualization = Turning raw data into visual charts and graphs to understand information quickly and make better decisions

# Data
- Data is a collection of raw facts, figures, observations, or values that can be processed to obtain meaningful information.

- Data is raw information that has not yet been analyzed or organized.

# Types of Qualitative Data
# A) Nominal Data

- Data represented by numbers.

# 04.Histogram
- A Histogram is used to show the distribution (frequency) of numerical data.

- It groups data into ranges called bins and shows how many values fall into each range.

# Why Use a Histogram?

- Understand data distribution

- Find data spread

- Detect skewness

- Identify outliers

- Check if data is normally distributed

| Histogram             | Bar Plot         |
| --------------------- | ---------------- |
| Numerical Data        | Categorical Data |
| Shows Distribution    | Shows Comparison |
| Bars Touch Each Other | Bars Have Gaps   |
| Uses Bins (Ranges)    | Uses Categories  |

# bins → Number of intervals/ranges

```python
# A Histogram is a graphical representation used to visualize the distribution of numerical data. It groups data into intervals called bins and 
# displays
# the frequency of values within each interval. It helps analyze data spread, patterns, skewness, and outliers.
```

# 05.Density Plot (KDE Plot)

- A Density Plot is a smooth version of a histogram that shows the probability distribution of data.

- Instead of bars, it uses a smooth curve to show where data points are concentrated.

# Why Use a Density Plot?

- Understand data distribution

- Compare multiple distributions

- Identify peaks in data

- Check normal distribution

- Smoother than a histogram

| Histogram       | Density Plot              |
| --------------- | ------------------------- |
| Uses bars       | Uses smooth curve         |
| Shows frequency | Shows probability density |
| Depends on bins | No bins required          |
| Less smooth     | More smooth               |


# When to Use Density Plot?
- To visualize data distribution
- To compare two datasets
- To check if data is normally distributed
- During Exploratory Data Analysis (EDA)

```python
# A Density Plot (KDE Plot) is a graphical representation used to visualize the probability distribution of continuous numerical data. It provides 
# a smooth curve that helps identify the shape, spread, peaks, and concentration of data more clearly than a histogram.
```

# Heat Map (Heatmap)
- A Heat Map is a data visualization technique that uses colors to represent the magnitude of values in a dataset.

# Why Use a Heat Map?
- Find patterns quickly
- Identify high and low values
- Show correlations between variables
- Analyze large datasets easily
- Useful in Data Science and Business Analytics 

# Advantages

- Easy to understand

- Quickly identifies patterns

- Good for large datasets

- Visually attractive

# Disadvantages

- Exact values can be hard to see without labels

- Too many colors may confuse users

- Not suitable for small datasets

# Box Plot (Box and Whisker Plot)

- A Box Plot is a statistical graph used to show the distribution of data, including:

- Minimum Value
- First Quartile (Q1)
- Median (Q2)
- Third Quartile (Q3)
- Maximum Value
- Outliers

# Advantages

- Easy to detect outliers

- Shows data spread

- Summarizes large datasets

- Compares multiple groups

# Disadvantages

- Doesn't show exact individual values

- Less detailed than histogram

- Can hide multimodal distributions

# Subplots
- Subplots allow you to display multiple plots (graphs) in a single figure/window

# Why Use Subplots?

- Suppose you want to compare:

- Line Plot
- Bar Plot
- Scatter Plot
- Histogram

- Instead of opening 4 separate windows, you can show all 4 plots in one figure using subplots.

# Advantages

- Compare multiple charts together

- Saves space

- Better data analysis

- Professional dashboards

- Easy reporting
